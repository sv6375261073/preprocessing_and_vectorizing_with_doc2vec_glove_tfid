```python
import pandas as pd
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
import lxml
```

```python
# cleaning html data 
def removeStopwords(text,lst):
    clean=''
    text=text.lower()
    for word in text.split():
        if word not in lst:
            clean=clean+word+' '
    return clean.strip()
def basicClean(text):
    soup = BeautifulSoup(text, 'lxml')
    text=soup.text
    text=soup.text.lower()
    text=re.sub("['\d','\t']",' ',text)
    text=repr(text)
    text=re.sub(r"[,”;“:#?_@(./)|!$%^&*\~`\[\]§\"']",' ',text)
    text=' '.join([word.replace('\\n',' ') if len(word)>15 else word.replace('\\n','') for word in text.split()])
    text=text.replace('  ',' ')
    text=text.replace('- ','')
    text=text.replace('™','')
    text=' '.join([word for word in text.split() if len(word)>1 or word in ['a','an']])
    return text

def lemmatize(words):
    if type(words) is list:   # checking  argument is list or text line
        pass
#         words=[set(words)]   # unique words in word_list
    else:
        words=words.split()
    lemmatizer = nltk.stem.WordNetLemmatizer()
    # lemmatization
    try:
        lemma_words=[lemmatizer.lemmatize(word.strip()) for word in words]
        lemma_text=' '.join(lemma_words)
    except LookupError:
        nltk.download('wordnet')   # download if not available 
        lemma_words=[lemmatizer.lemmatize(word.strip()) for word in words]
        lemma_text=' '.join(lemma_words)
    return lemma_text

try:
    stop=stopwords.words('English')
except Exception as e:
    nltk.download('stopwords')
    stop=stopwords.words('english')
userStopList=['fd','th','cir','et','al','llc','llp','mo','il','mr','ii','-wcg','iii','-cv-','rx','wspf',
             'np','ﬁif','ﬁmr','fia','wdoc','ex','rel']
stop.extend(userStopList)
```

```python
df=pd.read_csv('Law.csv')
feature=df.copy()
```

```python

feature.dropna(inplace=True)
feature.Text=feature.Text.apply(func=basicClean)
feature.Text=feature.Text.apply(lambda x:removeStopwords(x,stop))
feature.Text=feature.Text.apply(func=lemmatize)
# feature.Text=feature.Text.apply(func=conclude)
# df['UniqueWords']=df.Text.apply(lambda x:set(x.split()))
```
